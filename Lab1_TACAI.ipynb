```python
import glob
import cv2
import random as random
import matplotlib.pyplot as plt
import numpy as np

```

```python
base_path = "/content/drive/My Drive/TACAI/"
#take the path for all the jpg 
image_path_by_class =[]

for i in range(4):
  c_p = base_path + str(i) + "/*.jpg"
  image_path_by_class.append(glob.glob(c_p))

print(image_path_by_class)
```

```python
# shuffle all the images to be random 
for a in image_path_by_class:
  random.shuffle(a, lambda : .5)
print(image_path_by_class)
```

```python
# impartire date in train si test 70% 30%

train_lb = []
test_lb = []
for a in image_path_by_class:
  cls_t =[]
  cls_v = []
  for i,e in enumerate(a):
    if i< 7 :
      cls_t.append(e)
    else :
      cls_v.append(e)
  train_lb.append(cls_t)
  test_lb.append(cls_v)

print(train_lb)
print(test_lb)
```

```python
# incarcare imagini

train = []
test =[]

for cls in train_lb :
  tmp = []

  for e in cls :
    im = cv2.imread(e)
    tmp.append(im)
  train.append(tmp) 

for cls in test_lb :
  tmp = []

  for e in cls :
    im = cv2.imread(e)
    tmp.append(im)
  test.append(tmp) 
```

```python
# afisare imagini

plt.figure(), plt.imshow(train[0][0][...,::-1]) # ptc cv2 citeste BGR si imshow printeaza RGB
plt.figure(), plt.imshow(test[0][0][...,::-1])
```

```python
print(np.shape(test[0][0])) # imagini sunt mari pentru a lucra in timp 
```

```python
for i,cls in enumerate(train) :
  for j, im in enumerate(cls) :
    r = cv2.resize(im,(480,640))
    train[i][j] = r

for i,cls in enumerate(test) :
  for j, im in enumerate(cls) :
    r = cv2.resize(im,(480,640))
    test[i][j] = r
    
plt.figure(), plt.imshow(train[0][0][...,::-1]) # ptc cv2 citeste BGR si imshow printeaza RGB
plt.figure(), plt.imshow(test[0][0][...,::-1])
```

```python
for i,cls in enumerate(train) :
  for j, im in enumerate(cls) :
    fixed_height = 640
    height_percent = (fixed_height / float(im.shape[1]))
    width_size = int((float(im.shape[0]) * float(height_percent)))
    r = cv2.resize(im,(width_size,fixed_height)).astype('uint8')
    train[i][j] = r

for i,cls in enumerate(test) :
  for j, im in enumerate(cls) :
    fixed_height = 640
    height_percent = (fixed_height / float(im.shape[1]))
    width_size =np.int64((float(im.shape[0]) * float(height_percent)))
    r = cv2.resize(im,(width_size,fixed_height)).astype('uint8')
    test[i][j] = r
    
plt.figure(), plt.imshow(train[0][0][...,::-1]) # ptc cv2 citeste BGR si imshow printeaza RGB
plt.figure(), plt.imshow(test[0][0][...,::-1])
```

```python
#imag sunt BGR le fac RGB

img = train_g[0][5]
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(), plt.imshow(img)

# functia de detectie de piele de la laboratorul de IVOM 
def condition(img):
    r= img[:,:,0]
    g= img[:,:,1]
    b = img[:,:,2]
    C1 =(r>95) & (g>40) & (b>20)
    C2 =(np.maximum(np.maximum(r,g),b) - np.minimum(np.minimum(r,g),b))>15
    C3 = abs(r-g)>15
    C4 = (r>g)& (r>b)
    C5 = (r>220) & (g>210) & (b>170)
    C6 = abs(r-g)<= 15
    C7 = (r>b)& (g>b)
    cond1 = (C1*C2*C3*C4) + (C4*C5*C6*C7)
    cond1[cond1>1] = 1
    
    return cond1

masca = condition(img)
plt.figure(), plt.imshow(masca, cmap= 'gray')
```

```python
# sa gasesc bounding box pentru mana, prea multa informatie sus -> vreau mana sa ocupe cat mai mult din imagine

contours, hierarchy = cv2.findContours(masca.astype(np.uint8), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
img_copy = img.copy()
#cv2.drawContours(img_copy, contours, -1, (0,255,0), 3)

plt.figure(), plt.imshow(masca, cmap= 'gray')
plt.figure(), plt.imshow(img_copy, cmap= 'gray')

bounding_box =None
arie_max = 0
for contur in contours:
  #determin cel mai mare dreptnghi in functie dde aria calculata
  #x,y, w, h
  tmp = cv2.boundingRect(contur)
  if arie_max < tmp[2]*tmp[3]:
    arie_max = tmp[2]*tmp[3]
    bounding_box = tmp

cv2.rectangle(img_copy,(bounding_box[0],bounding_box[1]),(bounding_box[0]+bounding_box[2],bounding_box[1]+bounding_box[3]),(255,0,0),2)

plt.figure(), plt.imshow(img_copy, cmap= 'gray')

#acum ii dau crop in functie de cel mai mare BB si ii dau crop si mastii

img_copy = img_copy[bounding_box[1]:bounding_box[1]+bounding_box[3],bounding_box[0]:bounding_box[0]+bounding_box[2]]
masca_copy = masca.copy()
masca_copy = masca_copy[bounding_box[1]:bounding_box[1]+bounding_box[3],bounding_box[0]:bounding_box[0]+bounding_box[2]]
plt.figure(), plt.imshow(img_copy, cmap= 'gray')
plt.figure(), plt.imshow(masca_copy, cmap= 'gray')

#aplicare masca pe imagine
img_mana = img_copy.copy()
img_mana[masca_copy==0] =0

plt.figure(), plt.imshow(img_mana, cmap= 'gray')
```
